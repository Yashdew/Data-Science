```python
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
import folium as fo
```

```python
map = fo.Map()
```

```python
map
```

```python
vol = pd.read_csv('volcano.csv')
```

```python
vol.head()
```

```python
lat_vo=list(vol['Latitude'])
lon_vo=list(vol['Longitude'])
name_vo=list(vol['Name'])
```

```python
voll = fo.FeatureGroup(name='My name')
```

```python
for lat,lon,name1 in zip(lat_vo,lon_vo,name_vo):
    voll.add_child(fo.Marker(location=[lat,lon],
                  popup=name1,
                  icon = fo.Icon(color='red')))
```

```python
map.add_child(voll)
```
