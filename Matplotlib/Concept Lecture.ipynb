```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.show()
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x **2
```

```python
x
```

```python
y
```

```python
#functional 
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g')

plt.subplot(1,2,2)
plt.plot(x,y,'r')
```

```python
#OOP
fig = plt.figure()

axex = fig.add_axes([0.1,0.1,0.8,0.8])

axex.plot(x,y)
axex.set_xlabel('hori')
axex.set_ylabel('verti')
axex.set_title('head')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_xlabel('hori1')
axes2.set_xlabel('hori2')

axes1.set_ylabel('verti1')
axes2.set_ylabel('hori2')

axes1.set_title('head1')
axes2.set_title('head2')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#Figure Size and DPI
 
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(3,2))

#ax = fig.add_axes([0,0,1,1])
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_pic.png',dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title('Title')
ax.set_xlabel('X Lable')

ax.set_ylabel('Y Lable')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.plot(y,x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x square')

ax.plot(x,x**3,label='x cube')

ax.legend(loc=0)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange',linewidth = 10,alpha = 0.5)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange',lw = 10,linestyle = '--')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange',lw = 1,ls = '-',marker = '<',markersize = 10,markerfacecolor ='blue',markeredgewidth=3,markeredgecolor='red')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange',lw = 2,ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
